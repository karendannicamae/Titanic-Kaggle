```python
import pandas as pd
import numpy as np
import random

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Training Set

```python
df1 = pd.read_csv('train.csv')
```

```python
# observing the first 10 observation of the training set

df1.head(10)
```

```python
# counting the null values for each column

df1.isnull().sum()
```

```python
# creating a list of all the values in the Age column and removing the null values from the list

list_age = [x for x in df1.Age.tolist() if (np.isnan(x) == False)]
```

```python
# filling the null values with Age values from list_age

df1.Age.fillna(random.choice(list_age), inplace = True)
```

```python
# checking that there is no null value in the Age column

df1.Age.isnull().sum()
```

```python
df1.head()
```

```python
# dropping unwanted columns

df1.drop(columns = ['PassengerId', 'Name', 'Ticket', 'Cabin'], inplace = True)
```

```python
# creating dummy variables from categorical variables

df_train = pd.get_dummies(df1, columns = ['Sex', 'Embarked'], drop_first = True)
df_train
```

```python
# splitting the features and target variable

X_train = df_train.drop(columns = 'Survived')
y_train = df_train.Survived
```

```python
print(X_train.shape)
print(y_train.shape)
```

### Test Set

```python
df2 = pd.read_csv('test.csv')
```

```python
df2.head()
```

```python
df2.isnull().sum()
```

```python
# creating a list of all the values in the Age column and removing the null values from the list

list_age_test = [x for x in df2.Age.tolist() if (np.isnan(x) == False)]
```

```python
# filling the null values with Age values from list_age

df2.Age.fillna(random.choice(list_age_test), inplace = True)
```

```python
# checking if there is null

df2.Age.isnull().sum()
```

```python
# dropping unwanted columns

df2.drop(columns = ['PassengerId', 'Name', 'Ticket', 'Cabin'], inplace = True)
```

```python
# creating dummy variables from categorical variables

X_test = pd.get_dummies(df2, columns = ['Sex', 'Embarked'], drop_first = True)
X_test
```

```python
X_test[pd.isnull(X_test.Fare)]
```

```python
# filling the null value for column Fare

X_test.Fare.fillna(X_test.Fare.mean(), inplace = True)
```

### --- XGBoost ---

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(max_depth = 3, random_state = 5, learning_rate = 0.1)
```

```python
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)
```

```python
y_pred_xgb
```

```python
pred_xgb = pd.DataFrame(y_pred_xgb)
```

```python
pred_xgb.to_csv('pred_xgb.csv')
```

This had the accuracy of 0.76555.

### --- Random Forest ---

```python
# instantiating Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 10, max_depth = 6, min_samples_leaf = 2, random_state = 8)
```

```python
# training

rf.fit(X_train,y_train)
```

```python
# prediction

y_pred_rf = rf.predict(X_test)
```

```python
pred_rf = pd.DataFrame(y_pred_rf)
```

```python
pred_rf.to_csv('pred_rf.csv')
```

This had the accuracy of 0.79425
